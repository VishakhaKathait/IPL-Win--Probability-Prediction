```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv('matches.csv') #resd_csv is used to load a CSV file as a pandas datafram
delivery = pd.read_csv('deliveries.csv')
```

```python
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
#now we have to find total runs of every inning in every match so for doing this we are group by 
#reset_index = covert it into dataframe
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]#filtering the data
# we have to find the probility after 1 inning so we are taking total runs of 1st inning only.
```

```python
total_score_df
```

```python
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')#join condn
#now we merge the above datframe to the match dataframe 
#because in match dataframe we have every details except we dont have a record of total runs made by batting team in first inning.
```

```python
match_df['team1'].unique()
#get to know how many teams are playing
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
#team list
#removing teams which are not a part of ipl or their names are change   
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
#replacing names
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
#team 1 includes only those teams which are in the team list
#same for the team 2
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
#we are not taking the matches where duckworth lewis  is > 0
#we want data of those method where rain didnt affect
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
#batting team, bowling team=delivery dataframe
#city,total runs=match datframe
```

```python
delivery_df = match_df.merge(delivery,on='match_id') #on='match_id' ==joining condition
#match_df contains onl those coloumn which we want to display so we merge it with delivery dataframe because rest of data we need is in the delivery
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
#
```

```python
delivery_df
```

```python
#now we have batting  team,bowling team , city and total runs coloumn which get from delivery & match dataframe
# now main task to find the currentrunrate,runs_left,balls_left,wickts_left, requiredrunrate and result
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
# current score after each ball
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
#runsleft= total runs-current score (simple maths)
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
#balls left= how many balls are left after playing 1 ball in each over
#126-(6+1)= 119balls or 120 -1=119
```

```python
delivery_df
```

```python
#wickets left coloumn
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
#replacing player_dismissed nan value to 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
#if there is name of the player in player dismissed coloumn then replaceit with 1 otherwise with 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
#here we have converted the column into integer
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets #wickets left;
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
#now we will make a coloumn current runrate
#the first ball is 1/6th of an over 
#crr=runs/overs
delivery_df['crr']=(delivery_df['current_score'] *6)/(120-delivery_df['balls_left'])
#runs=current score
#we dont have over so we have a formula=(120-ballsleft)/6
```

```python
#required runrate coloumn
#runsleft/oversleft
delivery_df['rrr']=(delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
# result coloumn
#if the batting team and the winner team is same then the result should be 1 else 0
def result(row):
     return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
   final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df=final_df.sample(final_df.shape[0]) #shuffle the table
```

```python
final_df.sample() #randomly checking 
```

```python

final_df.dropna(inplace=True) #there are some value which are null
```

```python
final_df =  final_df[final_df['balls_left'] != 0]#we aretking only those values where balls_left is not equalto 0if we take then our min runrate and max runrate will be -infinite and infinite
```

```python
#now we are doing model building on this table which we have created i.e final_df
#first step is to train_test_split which we done from the sklearn
X = final_df.iloc[:,:-1]#all rows except last coloumn
y = final_df.iloc[:,-1]#all rows upto last coloumn
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
#herewe have 3 coloum which are string type so we are applying onehotencoding
#step 1
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
#step 2
#we are applying logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
#we make pipeline step 1 is tranformer & 2 step is logistic regression
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
#hereif we use randomforest instead of logistic regression the accuracy is much higher about 9.99 because here we are giving direct input
#if we want to show the probability then logisticregression gives good result rather than rf
```

```python
def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['total_runs_x']))
  
```

```python
#passing 3 hings (match complete data(each ball details),particular match ,giving model)
#after every over it calculate the run probability of  the chasing team  ang provide it in the dataframe
#total datframe mai se current match ke saare rows ko niklana hai or usko ek ek krke pipe mai bhejna hai
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
    
```

```python
temp_df,target = match_progression(delivery_df,74,pipe)
temp_df
```

```python
#take the above data and plot it in the graph
#blue bars represent at each over what are total run
# yellow line represent at each over what are total wickets
#green curve is wining probability
#red curve is losing probability
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.plot(temp_df['end_of_over'],temp_df['wickets_in_over'],color='yellow',linewidth=3)
plt.plot(temp_df['end_of_over'],temp_df['win'],color='#00a65a',linewidth=4)
plt.plot(temp_df['end_of_over'],temp_df['lose'],color='red',linewidth=4)
plt.bar(temp_df['end_of_over'],temp_df['runs_after_over'])
plt.title('Target-' + str(target))
```
